```python
import pandas as pd

df = pd.read_csv("googleplaystore.csv")

df.head(1000)
```

```python
three_col_df = df[['App','Category','Installs','Price']]
```

```python
three_col_df.head()
```

```python
def clean_installs(installs_num):

    installs_str = installs_num.replace("+", "").replace(",", "")
    
    try:
        installs = int(installs_str)
    except:
        installs =  0
    
    return installs

three_col_df['Installs'] = three_col_df.apply(lambda row: clean_installs(row['Installs']),  axis=1)

```

```python
three_col_df.head()
```

```python
result = three_col_df.sort_values(['Installs', 'Category'], ascending= False)
```
